```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
# functonal modal(graph based, branching , merging, skip) , subclassed model - all different models


from tensorflow.keras.layers import Dense, Flatten
# conv2d, dropout, lstm, - one lego block - all different layers
# Dense(128, activation='relu')

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Train
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# Evaluate
print("Test accuracy:", model.evaluate(x_test, y_test)[1])
```

```python
import numpy as np

# Pick one test image
idx = 0
image = x_test[idx]

# Model prediction (returns probabilities for 0–9)
pred = model.predict(image.reshape(1, 28, 28))  

# Show predicted label
print("Predicted digit:", np.argmax(pred))
print("Actual digit:", y_test[idx])
```
